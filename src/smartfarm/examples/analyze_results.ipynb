```python
# Import necessary modules
import matplotlib.pyplot as plt
import numpy as np
import os
```

## Analysis across runs

```python
ga_parallel_results_folder   = "ga_parallel_results"
values_costs_folder          = "unique_values_costs_results"
best_state_variables_folder  = "best_state_variable_results"
worst_state_variables_folder = "worst_state_variable_results"
```

## Revenue analysis

```python
fig, ax = plt.subplots(figsize=(8, 5))

for i in range(1, 11): # Runs 1 to 10

    values_path = os.path.join(
        ga_parallel_results_folder,
        values_costs_folder,
        f"ga_parallel_result_{i}_unique_values.csv"
    )

    costs_path = os.path.join(
        ga_parallel_results_folder,
        values_costs_folder,
        f"ga_parallel_result_{i}_unique_costs.csv"
    )

    with open(values_path, "rb") as f:
        unique_values = np.loadtxt(f, delimiter=",")

    with open(costs_path, "rb") as f:
        unique_costs = np.loadtxt(f, delimiter=",")
        unique_revenues = -unique_costs

    member_ids = np.arange(unique_revenues.shape[0])
    ax.plot(member_ids, unique_revenues[::-1], label=f"Run {i}")

ax.set_title("Revenues across final populations", fontsize=16, pad=15)
ax.set_xlabel("Member", fontsize=13)
ax.set_ylabel("Revenue ($/acre)", fontsize=13)
ax.legend()

fig.tight_layout()
```

## Growth analysis - Best in the final populations

```python
# Initialize the figure for plant growth analysis
fig, axs = plt.subplots(5, 1, figsize=(10, 15))

# Loop over all 10 of the best runs to analyze plant growth
for i in range(1, 11): # Runs 1 to 10

    state_variable_path = os.path.join(
        ga_parallel_results_folder,
        best_state_variables_folder, # <-- where we specify best or worst !!!
        f"state_variables_result_{i}.csv"
    )

    with open(state_variable_path, "rb") as f: # Load state variables but skip header
        state_variables = np.loadtxt(f, delimiter=",", skiprows=1)

    # Extract state variables: height (h), leaf area (A), number of leaves (N), spikelet count (c), fruit biomass (P)
    time = state_variables[:, 0]
    h = state_variables[:, 1]  # Plant height over time
    A = state_variables[:, 2]  # Leaf area over time
    N = state_variables[:, 3]  # Number of leaves over time
    c = state_variables[:, 4]  # Spikelet count over time
    P = state_variables[:, 5]  # Fruit biomass over time


    axs[0].plot(time, h, label=f'Run {i}')
    axs[1].plot(time, A, label=f'Run {i}')
    axs[2].plot(time, N, label=f'Run {i}')
    axs[3].plot(time, c, label=f'Run {i}')
    axs[4].plot(time, P, label=f'Run {i}')

# Plant Height
axs[0].set_xlabel('Time (hr)')
axs[0].set_ylabel('Plant Height (m)')
axs[0].set_title(f'Plant Height vs.Time')
axs[0].legend()
axs[0].grid(True)

# Leaf Area
axs[1].set_xlabel('Time (hr)')
axs[1].set_ylabel('Leaf Area (m)')
axs[1].set_title(f'Leaf Area vs.Time')
axs[1].legend()
axs[1].grid(True)

# Number of Leaves
axs[2].set_xlabel('Time (hr)')
axs[2].set_ylabel('Number of Leaves (unitless)')
axs[2].set_title(f'Number of Leaves vs.Time')
axs[2].legend()
axs[2].grid(True)

# Spikelet Count
axs[3].set_xlabel('Time (hr)')
axs[3].set_ylabel('Spikelet Count (unitless)')
axs[3].set_title(f'Spikelet Count vs.Time')
axs[3].legend()
axs[3].grid(True)

# Fruit Biomass
axs[4].set_xlabel('Time (hr)')
axs[4].set_ylabel('Fruit Biomass (kg)')
axs[4].set_title(f'Fruit Biomass vs.Time')
axs[4].legend()
axs[4].grid(True)

fig.suptitle(f'Hourly Plant Growth over Season - Best Inputs', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
